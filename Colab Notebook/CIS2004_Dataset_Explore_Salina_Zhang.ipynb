```python
import matplotlib.pyplot as plt

formula_number = df.groupby("Formula").size().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(formula_number.index, formula_number.values)
plt.title("Top 10 Most Frequent Organic Formulas")
plt.xlabel("Formula")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Scatter plot of the all molecules in the CSV file which ranked the melting points from lowest to highest

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Organic-Compounds-MeltingPoints.csv', encoding='latin1')

# Sort by melting point (assuming column name is 'Melting_Pt')
df_sorted = df.sort_values(by='Melting_Pt')

# Create a scatter plot without annotations
plt.figure(figsize=(10, 6))
x_values = range(len(df_sorted))  # Use index for x-axis
plt.scatter(x_values, df_sorted['Melting_Pt'], color='blue', s=10)  # Smaller points
plt.xlabel('Sorted Index (Lowest to Highest Melting Point)')
plt.ylabel('Melting Point (C)')
plt.title('Scatter Plot of Melting Points of Organic Molecules')

plt.tight_layout()
plt.show()
```

The following code only shows the lowest 20 melting points of molecules from the whole CSV file

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Organic-Compounds-MeltingPoints.csv', encoding='latin1')

# Sort and take top 20 lowest
df_subset = df.sort_values(by='Melting_Pt').head(20)

# Create a scatter plot with annotations
plt.figure(figsize=(10, 6))
x_values = range(len(df_subset))
plt.scatter(x_values, df_subset['Melting_Pt'], color='blue', s=50)
plt.xlabel('Sorted Index (Lowest Melting Points)')
plt.ylabel('Melting Point (C)')
plt.title('Scatter Plot of Top 20 Lowest Melting Points')

for i, row in enumerate(df_subset.itertuples()):
    plt.annotate(row.Name, (i, row.Melting_Pt), textcoords="offset points", xytext=(0,5), ha='center')

plt.tight_layout()
plt.show()
```
